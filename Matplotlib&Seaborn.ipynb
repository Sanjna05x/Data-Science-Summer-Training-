**Matplotlib**


Definition: Matplotlib is the foundational Python library for creating static, animated, and interactive visualizations — line charts, bar charts, scatter plots, histograms, etc. It gives you fine-grained, low-level control over every part of a plot (axes, ticks, colors, labels).

matplotlib and seaborn are the two core Python visualization libraries, and they work together.

```python
import matplotlib.pyplot as plt         #chart types
import numpy as np
plt.hist([[1,2,3,4],[1,4,9,16]])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.pie([1,2,3,4])
plt.show()
```

**Customizing plot**

```python
import matplotlib.pyplot as plt          #customizing plot

x = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
y = [100, 200, 150, 300, 250]

plt.plot(x, y, 'r--o' , # red, dashed line, circle markers
         linewidth=2,
         label='Revenue')

plt.title('Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend()
plt.grid(True)
plt.show()
```

**Multiple lines on plot**

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[3,6,9,12,15]
plt.plot(x, y1, 'g-s', label='Product A')    #multiple lines on one plot
plt.plot(x, y2, label='Product B')
plt.grid(True)
plt.legend()
plt.show()
```

**Multiple plots (Subplots)**

```python
import matplotlib.pyplot as plt
import numpy as np

x=([1,2,3,4])
y=([1,4,9,16])
fig, axs = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns

axs[0].plot(x, y)
axs[0].set_title('Line Chart')

axs[1].bar(x, y)
axs[1].set_title('Bar Chart')

plt.tight_layout()
plt.show()
```

**Working with pandas**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np                               #Working with pandas

# Create DataFrame
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr'],
    'Revenue': [1000, 1500, 2000, 2500],           #Dataframe
    'Sales': [10, 15, 20, 25],
    'Category': ['Watch', 'Shoes', 'Watch', 'Bag']
}

df = pd.DataFrame(data)

# Bar Plot
df.plot(x='Month', y='Revenue', kind='bar', color='green')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

# Pie Chart
df['Category'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Category Distribution")
plt.ylabel('')
plt.show()
```

**Seaborn**

Definition: Seaborn is built on top of matplotlib and is designed specifically for statistical data visualization. It works seamlessly with Pandas DataFrames, has better default styling out of the box, and makes complex plots (like distributions, correlations, and categorical comparisons) much easier to create with less code, work with large datasets.

```python
import seaborn as sns
import matplotlib.pyplot as plt          #Import libraries
import pandas as pd
df = pd.read_csv('ecommerce_sales_data (1).csv')       #Load dataset ----work with pandas large datasets
sns.lineplot(data=df, x='Sales', y='Profit',hue='Category', palette='viridis')
plt.title('Sales Report')
plt.grid(True)
sns.set_style("darkgrid")
plt.show()

sns.countplot(data=df, x='Category')
plt.title('Category Distribution')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
df.head()
df.describe()
df.info()
```

**Types of chart in matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
df = sns.load_dataset("tips")

# 1. Line Chart
plt.plot(x, y)
plt.title("Line Chart")
plt.show()

# 2. Bar Chart
plt.bar(['A', 'B', 'C'], [10, 20, 15])
plt.title("Bar Chart")
plt.show()

# 3. Horizontal Bar Chart
plt.barh(['A', 'B', 'C'], [10, 20, 15])
plt.title("Horizontal Bar Chart")
plt.show()

# 4. Scatter Plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()

# 5. Histogram
plt.hist(np.random.randn(1000), bins=30)
plt.title("Histogram")
plt.show()

# 6. Pie Chart
plt.pie([30, 20, 50], labels=['A', 'B', 'C'], autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()

# 7. Box Plot
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Box Plot")
plt.show()

# 8. Violin Plot
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Violin Plot")
plt.show()

# 9. KDE Plot
sns.kdeplot(df['total_bill'])
plt.title("KDE Plot")
plt.show()

# 10. Heatmap
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()

# 11. Pairplot
sns.pairplot(df, hue='sex')
plt.show()

# 12. Count Plot
sns.countplot(x='day', data=df)
plt.title("Count Plot")
plt.show()

# 13. Area Chart
plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.plot(x, y, color="slateblue")
plt.title("Area Chart")
plt.show()

# 14. Stacked Bar Chart
categories = ['A', 'B', 'C']
values1 = [10, 20, 15]
values2 = [5, 10, 8]
plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, bottom=values1, label='Group 2')
plt.title("Stacked Bar Chart")
plt.legend()
plt.show()

# 15. Bubble Chart
sizes = np.random.randint(20, 200, size=5)
plt.scatter(x, y, s=sizes, alpha=0.5, color='green')
plt.title("Bubble Chart")
plt.show()
```

**Types of charts in Seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = sns.load_dataset("tips")

# 1. Histogram / Distribution Plot
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title("Histogram")
plt.show()

# 2. KDE Plot (Density Plot)
sns.kdeplot(df['total_bill'])
plt.title("KDE Plot")
plt.show()

# 3. Box Plot
sns.boxplot(x='day', y='total_bill', data=df)
plt.title("Box Plot")
plt.show()

# 4. Violin Plot
sns.violinplot(x='day', y='total_bill', data=df)
plt.title("Violin Plot")
plt.show()

# 5. Scatter Plot
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex')
plt.title("Scatter Plot")
plt.show()

# 6. Line Plot
sns.lineplot(x='size', y='total_bill', data=df)
plt.title("Line Plot")
plt.show()

# 7. Bar Plot
sns.barplot(x='day', y='total_bill', data=df)
plt.title("Bar Plot")
plt.show()

# 8. Count Plot
sns.countplot(x='day', data=df)
plt.title("Count Plot")
plt.show()

# 9. Heatmap
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()

# 10. Pairplot
sns.pairplot(df, hue='sex')
plt.show()

# 11. Joint Plot
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
plt.show()

# 12. Strip Plot
sns.stripplot(x='day', y='total_bill', data=df, jitter=True)
plt.title("Strip Plot")
plt.show()

# 13. Swarm Plot
sns.swarmplot(x='day', y='total_bill', data=df)
plt.title("Swarm Plot")
plt.show()

# 14. Reg Plot (scatter + regression line)
sns.regplot(x='total_bill', y='tip', data=df)
plt.title("Reg Plot")
plt.show()

# 15. Cat Plot (flexible categorical plotting)
sns.catplot(x='day', y='total_bill', data=df, kind='box')
plt.show()

# 16. FacetGrid (multiple plots split by category)
g = sns.FacetGrid(df, col='sex')
g.map(sns.histplot, 'total_bill')
plt.show()
```
