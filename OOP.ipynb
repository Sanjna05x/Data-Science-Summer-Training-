**Object-Oriented Programming (OOP)** in Python is a programming paradigm that organizes code around classes and objects to represent real-world entities and behaviors. It breaks down complex systems into modular, reusable components

**1. Classes and Objects**

**Class:** A blueprint or template for creating objects. It defines data variables (attributes) and functions (methods).

**Object**: An instance of a class that contains real data and executes the methods.

**`__init__`:** The constructor method that executes automatically when a new object is initialized.

**self:** A reference pointer pointing directly to the current instance of the class.

```python
# Class Car (with __init__)

class Car:

    def __init__(self, brand):
        self.brand = brand

c1 = Car("BMW")
c2 = Car("Tesla")

print(c1.brand)
print(c2.brand)
```

```python
#Without _init_

class Student:

    def set_function(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"My name is {self.name}")
        print(f"My age is {self.age}")


s1 = Student()
s1.set_function("Sanjna", 20)
s1.display()
```

| With `__init__()`                             | Without `__init__()`                           |
| --------------------------------------------- | ---------------------------------------------- |
| Runs automatically when an object is created. | You must call another method manually.         |
| Initializes object data immediately.          | Object is empty until you set the data.        |
| Less code to use.                             | Requires an extra method call.                 |
| Preferred and most commonly used in Python.   | Used only in simple examples or special cases. |


**Multiple Objects**

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

e1 = Employee("Aman", 50000)
e2 = Employee("Riya", 60000)

print(e1.name, e1.salary)
print(e2.name, e2.salary)
```


**Calling One Method from Another**

```python
class Student:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print("Hello", self.name)

    def welcome(self):
        self.greet()
        print("Welcome to Python")

s1 = Student("Sanjna")
s1.welcome()
```

**Simple Calculator Using OOP**

```python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

calc = Calculator()

print(calc.add(10, 5))
print(calc.subtract(10, 5))
```

**Student Result Example**

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def result(self):
        if self.marks >= 40:
            print(self.name, "Pass")
        else:
            print(self.name, "Fail")

s1 = Student("Sanjna", 85)
s2 = Student("Rahul", 32)

s1.result()
s2.result()
```

**The Four Pillars of OOP**

**Inheritance** allows a new class (child/derived) to absorb attributes and methods from an existing class (parent/base). It eliminates redundant duplicate code.

*Use super()* to invoke parent methods directly

**1 Single Inheritance**

One child class inherits from one parent class.

```python
class Animal:
    def eat(self):
        print("Eating")

class Dog(Animal):
    def bark(self):
        print("Barking")

d = Dog()
d.eat()   # inherited
d.bark()
```

**2. Multiple Inheritance**

One child class inherits from more than one parent class.

```python
class Father:
    def skills(self):
        print("Gardening")

class Mother:
    def skills(self):
        print("Cooking")

class Child(Father, Mother):
    pass

c = Child()
c.skills()  # Gardening (Father comes first in MRO)
```

**3. Multilevel Inheritance**

A class inherits from a child class, forming a chain.

```python
class Animal:
    def eat(self):
        print("Eating...")

class Dog(Animal):
    def bark(self):
        print("Barking...")

class Puppy(Dog):
    def weep(self):
        print("Weeping...")

p = Puppy()
p.eat()   # from Animal
p.bark()  # from Dog
p.weep()  # own method
```

**4. Hierarchical Inheritance**

Multiple child classes inherit from the same parent class.

```python
class Animal:
    def eat(self):
        print("Eating...")

class Dog(Animal):
    def bark(self):
        print("Barking...")

class Cat(Animal):
    def meow(self):
        print("Meowing...")

# Creating objects (instances) of the classes
d = Dog()   # object of Dog
c = Cat()   # object of Cat

d.eat()     # inherited from Animal → Eating...
d.bark()    # Dog's own method → Barking...

c.eat()     # inherited from Animal → Eating...
c.meow()    # Cat's own method → Meowing...
```

**5. Hybrid Inheritance**

A combination of two or more types of inheritance (e.g., multiple + multilevel).

```python
class Animal:
    def eat(self):
        print("Eating...")

class Pet:
    def domesticated(self):
        print("Domesticated")

class Dog(Animal):
    def bark(self):
        print("Barking...")

class Puppy(Dog, Pet):  # multilevel + multiple
    def weep(self):
        print("Weeping...")

p = Puppy()
p.eat()
p.bark()
p.domesticated()
p.weep()
```

**Polymorphism**

Polymorphism allows different classes to share identical method names but carry unique execution behaviors.

Python achieves this natively through duck typing and method overriding and can't support method overloading .

Many forms example TV and AC can be on one button

**Polymorphism (Method Overriding)**

```python
class Animal:
    def sound(self):  
        print("Animal makes a sound")


class Dog(Animal):
    def sound(self): # same method name 
        print("Dog barks")


d = Dog()
d.sound()
```

**Abstraction:**

Abstraction hides low-level implementation details and reveals only vital functionalities.

Import the abc module to build abstract base classes.

Abstract methods act as strict structural interfaces that child subclasses must implement.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC): # Abstract Class
    @abstractmethod
    def start_engine(self):
        pass

class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started."
        
bike = Motorcycle()
print(bike.start_engine())
```

 **Encapsulation:**
 
 Encapsulation wraps data and functional logic within a singular unit while hiding internal access rules.
 
**Protected attributes:** Prefixed with a single underscore _ (indicates internal use).

**Private attributes:** Prefixed with a double underscore __ (triggers name mangling to prevent outside access)

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance # Private attribute

    # Getter method to access protected data safely
    def get_balance(self):
        return self.__balance

account = BankAccount("Alice", 5000)
print(account.get_balance()) 
```

**Decorator:**

A decorator in Python is a design pattern that allows you to modify or extend the behavior of a function, method, or class without permanently changing its source code.

```python
def decorator(func):

    def wrapper():
        print("Before function call")
        func()
        print("After function call")

    return wrapper


@decorator
def greet():
    print("Hello")


greet()
```

**Exception Handling in Python**

Exception handling is used to handle errors in a program so that it does not crash.

**try, except, and else**

The else block executes only if no exception occurs.

```python
try:
    num = int(input("Enter a number: "))
    print(100 / num)

except ZeroDivisionError:
    print("Cannot divide by zero.")

else:
    print("Program executed successfully.")
```
